```python
import torch
from torch.special import bessel_j1 as J1
from torch.special import bessel_j0 as J0
import matplotlib.pyplot as plt
import numpy as np

# Define the range of angles (theta) in radians
theta = torch.linspace(0, 2*np.pi, 100)

# Define the range of radial coordinates (r)
r = torch.linspace(0, 20, 100)

# Compute the Bessel functions J0 and J1
J0 = J0(r)
J1 = J1(r)

# Plot J0
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(r, J0, label='J0', color='b')
plt.title('Bessel Function J0')
plt.xlabel('r')
plt.ylabel('J0')
plt.grid(True)
plt.legend()

# Plot J1
plt.subplot(1, 2, 2)
plt.plot(r, J1, label='J1', color='r')
plt.title('Bessel Function J1')
plt.xlabel('r')
plt.ylabel('J1')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
